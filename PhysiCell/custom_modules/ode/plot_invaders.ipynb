```python
import tellurium as te
import matplotlib.pyplot as plt
import numpy as np
# from ipywidgets import *
# from math import pi
```

```python
agniSeirModel = '''
    // fluxes
    E1: S -> E; n * gamma1 * I * S/S0; //exposure
    E2: E -> I; n *  kInfection  * E; //infection
    E3: E -> S; n * beta * E; //exposure without infection
    E4: I -> R; mu * I; // recovery

    // parameters
    n = 10; 
    gamma1 = 0.1;
    kInfection = 0.6;
    beta = 0.05;
    mu = 1./5.;

    // Initial conditions
    S0 = 1E6;

    I = 1;
    E = 1;
    R = 0;
    S = S0 - I - R - E;
'''
m = te.loada(agniSeirModel)
m.simulate(0, 100, 300)
m.plot()
```

```python
invader_model = '''
    // fluxes
    E1:   -> B; 0.4 * B;
    E2: I -> ; 0.5 * B * I/I0;
    
    // Initial conditions
    I0 = 100;

    I = I0;
    B = 1;
'''
m = te.loada(invader_model)
m.simulate(0, 2000, 1000)
m.plot()
```
